```python
traindatanp = np.array(traindatanp,dtype=np.float64)
meanVal = np.mean(traindatanp[:,range(6)],axis=0)
stdVal = np.std(traindatanp[:,range(6)],axis=0)
print(stdVal)
meanVal
```

```python
traindatanp[:,range(6)] = (traindatanp[:,range(6)]-meanVal)/stdVal
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(traindatanp[:,6:-1])
```

```python
traindatanp_a = np.array(traindatanp[:,:6])
traindatanp_c = np.array(traindatanp[:,-1])
traindatanp_c = traindatanp_c.reshape((traindatanp_c.shape[0],1))
traindatanp_b = np.array(enc.transform(traindatanp[:,6:-1]).toarray())
```

```python
#print(traindatanp_a.shape)
#print(traindatanp_b.shape)
#print(traindatanp_c.shape)
traindatanp = np.concatenate((traindatanp_a,traindatanp_b,traindatanp_c),axis=1)
```

```python
testdatanp_a = np.array(testdatanp[:,:6])
testdatanp_b = np.array(enc.transform(testdatanp[:,6:]).toarray())
```

```python
testdatanp = np.concatenate((testdatanp_a,testdatanp_b),axis=1)
```

```python
N,M = traindatanp.shape
M = M-1
N
```

```python
np.random.shuffle(traindatanp)
num = 335000
X_train = traindatanp[:num,:-1]
y_train = np.array(traindatanp[:num,-1],dtype=np.int64)
#print(X_train[0], y_train[0])
X_cv = traindatanp[num:,:-1]
y_cv = np.array(traindatanp[num:,-1],dtype=np.int64)
#print(X_cv[0], y_cv[0])
```

```python
import torch 
import torch.nn as nn
```

```python
model = nn.Sequential(nn.Linear(M,int(M/2),True),
                      nn.Tanh(),
                      nn.Linear(int(M/2),int(M/4),True),
                      nn.ReLU(),
                      nn.Linear(int(M/4),2,True),
                     nn.Tanh())
dType = torch.FloatTensor
lossFunc = nn.CrossEntropyLoss().type(dType)
optimizer = torch.optim.Adam(model.parameters(),lr = 1e-2)
```

```python
X_train_torch = torch.FloatTensor(X_train)
X_train_var = torch.autograd.Variable(X_train_torch)
ans = model(X_train_var)
```

```python
y_train_torch = torch.LongTensor(y_train)
y_train_var = torch.autograd.Variable(y_train_torch)
```

```python
epochs = 1500
```

```python
for epoch in range(epochs):
    optimizer.zero_grad()
    forward = model(X_train_var)
    loss = lossFunc(forward,y_train_var)
    loss.backward()
    optimizer.step()
    
    print("epoch ",epoch, " loss = ", loss.data[0])
```

```python
X_cv_torch = torch.FloatTensor(X_cv)
X_cv_var = torch.autograd.Variable(X_cv_torch)
y_cv_torch = torch.LongTensor(y_cv)
```

```python
out_cv = model(X_cv_var)
out_cv_np = np.array(out_cv.data)
out_cv_np = (out_cv_np + 1)/2
```

```python
from sklearn import metrics
```

```python
score = metrics.roc_auc_score(y_cv,out_cv_np[:,1])
score
```
