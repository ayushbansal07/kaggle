```python
import numpy as np
import pandas as pd
```

```python
traindata = pd.read_csv('train.csv',index_col=0)
testdata = pd.read_csv('test.csv',index_col=0)
```

```python
traindata = traindata.drop(['desk_id','office_id','indicator_code','hedge_value','status'],axis=1)
traindata.head()
```

```python
testdata = testdata.drop(['desk_id','office_id','indicator_code','hedge_value','status'],axis=1)
testdata.head()
```

```python
traindata = np.array(traindata)
testdata = np.array(testdata)
```

```python
for data in traindata:
    data[0] = ord(data[0]) - ord('A')
```

```python
for data in testdata:
    data[0] = ord(data[0]) - ord('A')
```

```python
country_map = {}
for data in traindata:
    if data[3] not in country_map:
        country_map[data[3]] = len(country_map)
    data[3] = country_map[data[3]]
        
print(country_map)
```

```python
for data in testdata:
    data[3] = country_map[data[3]]
```

```python
currency_map = {}
for data in traindata:
    if data[5] not in currency_map:
        currency_map[data[5]] = len(currency_map)
    data[5] = currency_map[data[5]]
        
currency_map
```

```python
for data in testdata:
    data[5] = currency_map[data[5]]
```

```python
N,M = traindata.shape
M
```

```python
listOfIndices = [1,2,4,6,7,8,9]
meanVal = np.nanmean(traindata[:,listOfIndices],axis=0)
for data in traindata:
    for j in [2,6,7]:
        if np.isnan(data[j]):
            if j==2:
                data[j] = meanVal[1]
            elif j==6:
                data[j] = meanVal[3]
            else:
                data[j] = meanVal[4]
traindata = np.array(traindata,dtype=np.float64)
meanVal = np.mean(traindata[:,listOfIndices],axis=0)
stdVal = np.std(traindata[:,listOfIndices],axis=0)
```

```python
for data in traindata:
    for k in range(len(listOfIndices)):
        j = listOfIndices[k]
        data[j] = (data[j]-meanVal[k])/stdVal[k]  
```

```python
testdata = np.array(testdata,dtype=np.float64)
meantest = np.nanmean(testdata[:,6],axis=0)
for data in testdata:
    if np.isnan(data[6]):
        data[6] = meantest
testdata = np.array(testdata,dtype=np.float64)
```

```python
for data in testdata:
    for k in range(len(listOfIndices)):
        j = listOfIndices[k]
        data[j] = (data[j]-meanVal[k])/stdVal[k]     
```

```python
N,M
np.random.shuffle(traindata)
```

```python
num = 8500
X_train = traindata[:num,:-1]
y_train = traindata[:num,-1]
print(X_train[0], y_train[0])
X_cv = traindata[num:,:-1]
y_cv = traindata[num:,-1]
print(X_cv[0], y_cv[0])
```

```python
M = 11
import torch
import torch.nn as nn
```

```python
model = nn.Sequential(
    nn.Linear(M,2*M,True),
    nn.ReLU(),
    nn.Linear(2*M,4*M,True),
    nn.ReLU(),
    nn.BatchNorm1d(4*M),
    nn.Linear(4*M,2*M,True),
    nn.ReLU(),
    nn.Linear(2*M,M,True),
    nn.ReLU(),
    nn.Linear(M,1,True),
    nn.BatchNorm1d(1))
dType = torch.FloatTensor
lossFunc = torch.nn.MSELoss().type(dType)
optimizer = torch.optim.Adam(model.parameters(),lr = 1e-2)
```

```python
X_train_torch = torch.FloatTensor(X_train)
X_train_var = torch.autograd.Variable(X_train_torch)
ans = model(X_train_var)
```

```python
y_train_torch = torch.FloatTensor(y_train)
y_train_var = torch.autograd.Variable(y_train_torch)
```

```python
epochs = 5000
```

```python
for epoch in range(epochs):
    optimizer.zero_grad()
    forward = model(X_train_var)
    loss = lossFunc(forward,y_train_var)
    loss.backward()
    optimizer.step()
    
    print("epoch ",epoch, " loss = ", loss.data[0])
```

```python
X_cv_torch = torch.FloatTensor(X_cv)
X_cv_var = torch.autograd.Variable(X_cv_torch)
y_cv_torch = torch.FloatTensor(y_cv)
```

```python
out_cv = model(X_cv_var)
out_cv_np = np.array(out_cv.data)
out_cv_np 
```

```python
from sklearn import metrics
```

```python
score = metrics.r2_score(y_cv,out_cv_np)
score
```
